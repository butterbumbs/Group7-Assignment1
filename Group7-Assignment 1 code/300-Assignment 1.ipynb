```python
import nltk
import numpy
import matplotlib
import os
```

```python
FilmTokens = nltk.word_tokenize(Film)
```

```python
len(FilmTokens)
```

```python
print(FilmTokens)
```

```python
def lexical_diversity(FilmTokens):
    return len(set(FilmTokens)) / len(FilmTokens)
```

```python
lexical_diversity(FilmTokens)
```

```python
FilmFreq = nltk.FreqDist(FilmTokens)

FilmFreq.most_common(10)
```

```python
LongWords = []

for word in FilmTokens:
    if len(word) >= 10:
        LongWords.append(word)
        
LongWord_dist = nltk.FreqDist(LongWords)
for word, frequency in LongWord_dist.items():
    print(word, frequency)
```

```python
FilmSentences = nltk.sent_tokenize(Film)

LongestSentence = (max(FilmSentences, key=len))

print(LongestSentence)
print("Count:", len(LongestSentence.split()))
```

```python
from nltk.stem import PorterStemmer
```

```python
Longest_Sentence_Token = nltk.word_tokenize(LongestSentence)
print(Longest_Sentence_Token)
```

```python
Filmstemmer = PorterStemmer()

Film_Sentences_Stem = []

for word in Longest_Sentence_Token:
    Film_Sentences_Stem.append(Filmstemmer.stem(word))
Film_Sentences_Stem = " ". join(Film_Sentences_Stem)
print(Film_Sentences_Stem)
```
